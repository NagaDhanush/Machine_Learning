```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```
x = np.resize(np.linspace(-6, 10, 25), (30, 1))

noise = np.random.normal(loc=0, scale=0.2, size=(30, 1))

y= np.sin(x) + noise

plt.plot(x, y, 'r.')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```
print(x)
print(y)
```

```
model = LinearRegression()
```

```
model.fit(x, y)

y_hat = model.predict(x)
def plot_model(x, y, y_hat):
  plt.plot(x, y_hat, 'b-')
  plt.plot(x, y, 'r.')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid()
  plt.show()

plot_model(x, y, y_hat)  
```

Degree2 polynomial

```
poly = PolynomialFeatures(degree = 2,include_bias=False)
poly = poly.fit(x)
x_2 = poly.transform(x)

print(x_2)

```

```
model = LinearRegression()
model.fit(x_2, y)
y_hat = model.predict(x_2)
plot_model(x, y, y_hat)
```

```
poly = PolynomialFeatures(degree = 4,include_bias=False)
poly = poly.fit(x)
x_4 = poly.transform(x)

print(x_4)
```

```
model = LinearRegression()
model.fit(x_4, y)
y_hat = model.predict(x_4)
plot_model(x, y, y_hat)
```
